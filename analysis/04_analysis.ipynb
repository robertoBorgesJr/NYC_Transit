```python
import plotly.express as px

# Lendo apenas o sumário 
pdf_share = spark.table(f"{catalog}.{schema}.gold_market_share_summary").orderBy("reference_date").toPandas()

# Vamos ordenar as categorias para que o Yellow Taxi fique sempre na mesma posição
pdf_share = pdf_share.sort_values(by=["reference_date", "taxi_type"])

# Gerando o gráfico com a tabela agregada
fig = px.area(pdf_share, 
              x="reference_date", 
              y="market_share_pct", 
              color="taxi_type",
              category_orders={"taxi_type": ["App (Uber/Lyft)", "Yellow Taxi"]},
              title="Market Share: A Ascensão dos Apps em NYC (2009-2019)",
              labels={"market_share_pct": "Fatia de Mercado (%)", "reference_date": "Linha do Tempo"},
              color_discrete_map={"Yellow Taxi": "#FFD700", "App (Uber/Lyft)": "#000000"},
              template="plotly_dark")

fig.update_layout(
    yaxis_range=[0, 100],
    hovermode="x unified" # Mostra os dois valores ao passar o mouse
)
fig.show()
```
